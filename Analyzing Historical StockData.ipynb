```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Step 1: Fetch the webpage and parse it
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")

# Step 2: Find the correct table
tables = soup.find_all("tbody")
tesla_table = tables[1]  # 2nd table contains Tesla's quarterly revenue

# Step 3–6: Extract data
data = []
for row in tesla_table.find_all("tr"):
    cols = row.find_all("td")
    if len(cols) == 2:
        date = cols[0].text.strip()
        revenue = cols[1].text.strip().replace(",", "").replace("$", "")
        if revenue != "":  # Skip empty values
            data.append([date, revenue])

# Step 7: Create DataFrame
tesla_revenue = pd.DataFrame(data, columns=["Date", "Revenue"])

# Display result
print(tesla_revenue.head())
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Step 1: Download the HTML page
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
response = requests.get(url)
html_data_2 = response.text

# Step 2: Parse the HTML using BeautifulSoup
soup_2 = BeautifulSoup(html_data_2, "html.parser")

# Step 3: Locate the revenue table (2nd <tbody>)
gme_revenue_table = soup_2.find_all("tbody")[1]

# Step 4: Initialize empty DataFrame
gme_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Step 5: Loop through each row and extract Date and Revenue
for row in gme_revenue_table.find_all("tr"):
    cols = row.find_all("td")
    if len(cols) == 2:
        date = cols[0].text.strip()
        revenue = cols[1].text.strip()
        
        # Only add rows with valid revenue data
        if revenue != "-" and revenue != "":
            # Clean Revenue: remove $ and commas
            revenue_cleaned = revenue.replace("$", "").replace(",", "")
            
            # Append to DataFrame
            gme_revenue = pd.concat([gme_revenue, pd.DataFrame([[date, revenue_cleaned]], columns=["Date", "Revenue"])], ignore_index=True)

# Optional: Convert Revenue column to float
gme_revenue["Revenue"] = gme_revenue["Revenue"].astype(float)

# Step 6: Display the first few rows
print(gme_revenue.head())
```

```python
# Display the last five rows of gme_revenue
print(gme_revenue.tail())
```

Question 5: Plot Tesla Stock Graph

Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

Upload a screenshot of your results.

```python
import pandas as pd

# Ensure 'Date' columns are in datetime format
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'])
tesla_revenue['Date'] = pd.to_datetime(tesla_revenue['Date'])
```

```python
import matplotlib.pyplot as plt

def make_graph(stock_data, revenue_data, company):
    fig, ax1 = plt.subplots(figsize=(14, 6))

    # Plot stock price
    ax1.plot(stock_data['Date'], stock_data['Close'], color='blue', label='Stock Price')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Stock Price', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Create second y-axis for revenue
    ax2 = ax1.twinx()
    ax2.plot(revenue_data['Date'], revenue_data['Revenue'], color='red', label='Revenue')
    ax2.set_ylabel('Revenue', color='red')
    ax2.tick_params(axis='y', labelcolor='red')

    plt.title(f'{company} Stock Price and Revenue (up to June 2021)')
    fig.tight_layout()
    plt.show()
```

```python
make_graph(tesla_data, tesla_revenue, 'Tesla')
```
